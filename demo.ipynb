## Step 2: Download Sample PDFs

```python
import os
import urllib.request
from pathlib import Path

os.makedirs("pdfs", exist_ok=True)

# Download "Attention Is All You Need" paper (academic, multi-column)
url = "https://arxiv.org/pdf/1706.03762"
path = "pdfs/attention.pdf"
if not Path(path).exists():
    urllib.request.urlretrieve(url, path)
    print(f"Downloaded: {path}")
else:
    print(f"Already exists: {path}")

print(f"File size: {Path(path).stat().st_size / 1024:.1f} KB")
```

## Step 3: Convert with Auto Mode (Smart Routing)

```python
from pdf_pilot.convert import convert
import time

t0 = time.time()
doc = convert("pdfs/attention.pdf", "output_auto.md", engine="auto")
elapsed = time.time() - t0

print(f"Pages: {doc.page_count}")
print(f"Engine: {doc.metadata['engine']}")
print(f"Time: {elapsed:.1f}s")
print(f"Content length: {len(doc.raw_markdown)} characters")
print(f"Blocks: {len(doc.blocks)}")
```

```python
# Preview the first 500 characters
print(doc.raw_markdown[:500])
```

## Step 4: Compare Engines

Let's try PyMuPDF (fast) vs Docling (better for complex layouts):

```python
# PyMuPDF — fast mode
t0 = time.time()
doc_pymupdf = convert("pdfs/attention.pdf", engine="pymupdf")
time_pymupdf = time.time() - t0

print(f"PyMuPDF: {len(doc_pymupdf.raw_markdown)} chars, {time_pymupdf:.1f}s")
```

```python
# Docling — quality mode
t0 = time.time()
doc_docling = convert("pdfs/attention.pdf", engine="docling")
time_docling = time.time() - t0

print(f"Docling: {len(doc_docling.raw_markdown)} chars, {time_docling:.1f}s")
```

```python
# Comparison
print("=" * 60)
print(f"{'Engine':<15} {'Chars':<10} {'Time':<10} {'Blocks':<10}")
print("-" * 60)
print(f"{'PyMuPDF':<15} {len(doc_pymupdf.raw_markdown):<10} {time_pymupdf:<10.1f} {len(doc_pymupdf.blocks):<10}")
print(f"{'Docling':<15} {len(doc_docling.raw_markdown):<10} {time_docling:<10.1f} {len(doc_docling.blocks):<10}")
```

## Step 5: Word Output

```python
from pdf_pilot.convert import convert

doc = convert("pdfs/attention.pdf", "output.docx", engine="pymupdf")
print(f"Word document created: output.docx ({Path('output.docx').stat().st_size / 1024:.1f} KB)")

# Download the file
from google.colab import files
files.download("output.docx")
```

## Step 6: Upload Your Own PDF

Upload a PDF from your computer and convert it:

```python
from google.colab import files
from pdf_pilot.convert import convert

print("Upload a PDF file:")
uploaded = files.upload()

for filename in uploaded.keys():
    print(f"\nConverting: {filename}")
    doc = convert(filename, f"{filename}.md", engine="auto")
    print(f"Pages: {doc.page_count}, Engine: {doc.metadata['engine']}")
    print(f"Content: {len(doc.raw_markdown)} characters, {len(doc.blocks)} blocks")
    files.download(f"{filename}.md")
```
